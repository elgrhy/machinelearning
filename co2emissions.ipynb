After importing the dependencies, understading and importing the data

```python
df = pd.read_csv('/kaggle/input/co2-emission-by-vehicles/CO2 Emissions_Canada.csv')
```

```python
df.head()
```

Splitting the data

```python
X = df.drop('CO2 Emissions(g/km)', axis=1)
y = df['CO2 Emissions(g/km)']
```

```python
numerical_cols = df.drop('CO2 Emissions(g/km)', axis=1).select_dtypes(exclude='object').columns
numerical_transformer = StandardScaler()

categorical_cols = df.select_dtypes('object').columns
categorical_transfortmer = OneHotEncoder()

```

Preprocessor

```python
preprocessor = ColumnTransformer(
    transformers = [
        ('num', numerical_transformer, numerical_cols),
        ('cat', categorical_transfortmer, categorical_cols)
    ]
)
```

Model

```python
model = RandomForestRegressor(n_estimators=100, random_state=0)
```

```python
pipeline = make_pipeline(preprocessor, model)
pipeline.fit(X, y)
```

```python
pipeline.predict(X)
```

```python
with open('pipeline.pkl', 'wb') as f:
    pickle.dump(pipeline, f)
```
